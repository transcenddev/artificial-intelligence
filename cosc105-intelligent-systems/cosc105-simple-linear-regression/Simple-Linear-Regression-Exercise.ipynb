```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load the data
data = pd.read_csv('/content/sample_data/real_estate_price_size.csv')

# Define X (independent variable) and y (dependent variable)
X = data[['size']]  # Independent variable (size as DataFrame)
y = data['price']   # Dependent variable (price as Series)

# Add a constant to the independent variable (for the intercept term)
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary of the regression
print(results.summary())

# R-squared from the model (indicates the accuracy)
r_squared = results.rsquared
print(f"R-squared: {r_squared}")

# Predicted values (y_hat)
y_hat = results.predict(X)

# Residuals (differences between actual and predicted values)
residuals = y - y_hat

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(residuals**2))
print(f"RMSE: {rmse}")
```
